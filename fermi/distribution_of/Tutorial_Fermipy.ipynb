## Intro

Fermipy was developed at NASA Goddard (mainly) by Matt Wood and other people. Fermipy takes a configuration file in the YAML format, and then uses this to run the whole analysis chain along with optimized fitting routines to make your analysis significantly easier.

```python
from fermipy.gtanalysis import GTAnalysis
# Create an analysis object from the config file
for i in range(1,3):
    gta = GTAnalysis('config.yaml', logging={'verbosity': 3}, fileio={'outdir': 'J1928_2020-08-01'})
    'string%s' % i
```

We can normally ignore any 'warning' messages, they generally benign. Now we'll run our setup

```python
gta.setup()
```

Based on our defined parameters in the config file, the setup has ran gtselect, and filtered our photon files in the same way that the command line tool would have with gtselect and gtmktime. It has then created a livetime cube, using gtltcube. It then runs gtbin to bin our photons, gtexpcube2 to compute exposure, and gtsrcmaps, to generate a source map of our ROI. 

This is all very useful, because this allows us to run all of the science tools (gtselect, gtmktime, gtltcube, gtexpcube2, ...), in one command, based on our input from the config file. You can read in the boix above "Running gtselect", "Finished gtselect", and then "Running gtmktim", ... and so on.
It saves us time and effort, and reduces human errors ! 

## Plots

```python
from gammapy.maps import Map, MapAxis, WcsGeom

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc

from astropy.io import fits
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
## Location of your Fermi maps
path = "./J1928_2020-08-01"

## CCUBE map
map_name = "ccube_00"
fermi_map = Map.read("%s/%s.fits" % (path, map_name), 0)

# Interactive plot (as a fonction of the energy)
rc_params = {
    "figure.figsize": (7, 7),
    "font.size": 12,
    "axes.formatter.limits": (2, -2),
}

fermi_map.plot_interactive(add_cbar=True, stretch="sqrt", rc_params=rc_params, cmap='inferno')
```

```python
## Stack all the CCUBE maps to plot a CMAP map
stacked_map = fermi_map.sum_over_axes()

position = SkyCoord(292.292, 18.36, frame="icrs", unit="deg")
stacked_map_cutout = stacked_map.cutout(position=position, width=(6 * u.deg, 6 * u.deg))

#Plots
with plt.rc_context({'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'white'}):
    fig = plt.figure(figsize=(7, 7))
    stacked_map_cutout.smooth("0.1 deg").plot(add_cbar=True, stretch="log", vmin=0, vmax=10, cmap='inferno');
    fig.text(0.8, 0.9, 'Counts', ha='center', va='center')
    fig.text(0.45, 0.15, 'Right ascension', ha='center', va='center')
    fig.text(0.05, 0.5, 'Declination', ha='center', va='center', rotation=90)
    plt.xlabel(" ")
    plt.ylabel(" ")
    plt.savefig("%s/%s.png" % (path, map_name))
    plt.show()
```
